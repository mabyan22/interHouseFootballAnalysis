```python
import pandas as pd
import matplotlib.pyplot as plt
from soccerplots.radar_chart import Radar
```

```python
df = pd.read_csv('playerStatistics.csv')
df
```

```python
#creating parameters and ranges for radar
params=list(df.columns)
params=params[10:]
ranges=[]
player_values=dict()

for i in params:
    a=min(df[params][i])
    a=a-(a*0.25)
    b=max(df[params][i])
    b=b+(b*0.25)
    ranges.append((a,b))
    
for i in range(len(df['PlayerID'])):
    player_values[i]=df.iloc[i].values.tolist() 
    player_values[i]=player_values[i][10:]
```

```python
for i in range(len(df['PlayerID'])):
    radar = Radar()
    fig,ax=radar.plot_radar(
        ranges=ranges,params=params,values=player_values[i],
        radar_color=['indigo','violet'],alphas=[.25,.2],title={'title_name':df['PlayerID'][i]}
    )
```
